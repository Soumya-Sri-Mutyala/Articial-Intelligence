```python
import numpy as np
import pandas as pd
import matplotlib as plt
import random
import math
```

```python
#Genetic Algorithm variables

N=8     #Size of the chromosome population
pc=0.7  #Crossover probability
pm=0.01 #Mutation probability
no_generations = 200  # total number of generations
```

```python
# Converting Binary to Decimal

def binToDec(bin):
    return int(bin,2) 
```

```python
# Calculating the fitness function

def fitnessfn(x,y):
    fn = ((1-x)**2) * (math.exp(-(x**2)-((y+1)**2))) - (x-(x**3)-(y**3)) * (math.exp((-x**2)-(y**2)))
    return fn
```

```python
for z in range(200):
    print('Iteration'+'-' + str(z) + ':')
    print('-------------')
    array1 = []
    array2 = []

    #Generating random chromosomes(binary digits)
    for i in range(1,N+1):
        def chromosome(genes):
            key1=""
            for i in range(genes):
                temp=str(random.randint(0,1))
                key1 += temp
            return(key1)

        total_genes=16
        chromo=chromosome(total_genes)
        array2.append(chromo)

        # Dividing 16 bit string into two equal 8 bit strings
        X, Y = chromo[0:len(chromo)//2], chromo[len(chromo)//2:]

        # Converting two 8 bit strings into decimal
        X1Dec = binToDec(X)
        Y1Dec = binToDec(Y)

        #Calculating actual value of X and Y in the range of -2 and 2
        X2 = round(((X1Dec)*(4/255)-2),5)
        Y2 = round(((Y1Dec)*(4/255)-2),5)
        print(f'The values of X{i} and Y{i} are {X2},{Y2}')

        #Calculating the fitness value
        fit = fitnessfn(X2,Y2)
        fit = round(fit,5)
        print(f'Fitness value{i} is: {fit}')
        array1.append(fit)

        i += 1
    
    
    # Calculating Shifted f

    array3=[]


    for i in array1:
        if(i<0):
            shifted=True
        else:
            shifted=False

    if(shifted==True):
        total_fitness = 0
        for i in array1:
            i = i + 0.10000   # Shifted f(to make f all positive)
            total_fitness += i
        total_fitness=round(total_fitness,5) 
        print(f'Total fitness = {total_fitness}')

        for i in array1:
            i = i + 0.10000 # Shifted f(to make f all positive)
            fitness_ratio = round(((i/total_fitness)*100),5)
            array3.append(fitness_ratio)

    if(shifted==False):
        total_fitness = 0
        for i in array1:
            total_fitness += i
        total_fitness=round(total_fitness,5) 
        print(f'Total fitness = {total_fitness}')
        for i in array1:
            fitness_ratio = round(((i/total_fitness)*100),5)
            array3.append(fitness_ratio)
    
    
    # Roulette wheel method

    array5=[]
    array4=[]

    for x in range(0,N):
        value = random.randint(0,100)
        sum=array3[0]+array3[1]
        sum1=sum+array3[2]
        sum2=sum1+array3[3]
        sum3=sum2+array3[4]
        sum4=sum3+array3[5]
        sum5=sum4+array3[6]
        sum6=sum5+array3[7]
        if(0<value<array3[0]):
            var = 0
        elif(array3[0]<value<sum):
            var = 1
        elif(sum<value<sum1):
            var = 2
        elif(sum1<value<sum2):
            var = 3
        elif(sum2<value<sum3):
            var = 4
        elif(sum3<value<sum4):
            var = 5
        elif(sum4<value<sum5):
            var = 6
        elif(sum5<value<sum6):
            var = 7


        array4.append(array3[var])
        array5.append(array2[var])    


        #Crossover

    for i in range(1,N+1):
        crossover_value = np.random.uniform(0,1)
        k=np.random.randint(1,7)  # generate slicing value
        max_value = max(array4)   # max fitness ratio value
        max_index = array4.index(max_value)  # index of fitness ratio
        binvalue = array5[max_index]     # binary value to that fitness ratio
        X3, Y3 = binvalue[0:len(binvalue)//2], binvalue[len(binvalue)//2:]  # dividing binary value into two equal parts
        # slicing X and Y using k
        p11 = X3[0:k]   
        p12 = X3[k:8]

        p21 = Y3[0:k]
        p22 = Y3[k:8]
        if(crossover_value<pc):   
            # Concatenating the sliced values : Offsprings
            offspring1 = p11 + p22
            offspring2 = p21 + p12


        else:
            # Concatenating the sliced values : Offsprings
            offspring1 = p11 + p12
            offspring2 = p21 + p22

        print(f'After crossover, Offspring1 is: {offspring1}')
        print(f'After crossover, Offspring2 is: {offspring2}')

        # Mutation

        mutation_value = np.random.uniform(0,100)

        if(0<mutation_value<1):
            digits = [str(x) for x in str(offspring1)]
            digits1 = [str(x) for x in str(offspring2)]
            s=np.random.randint(0,8)

            #Mutation process of offspring1
            if(digits[s]=='0'):
                digits[s]='1'
            else:
                digits[s]='0'

            mute = ''.join([str(elem) for elem in digits])

            #Mutation process of offspring2
            if(digits1[s]=='0'):
                digits1[s]='1'
            else:
                digits1[s]='0'

            mute1 = ''.join([str(elem) for elem in digits1])

            chromosome_final = mute + mute1
            print(f'After Mutation, new chromosome{i}: {chromosome_final}')  

        else:
            chromosome_final = offspring1 + offspring2
            print(f'After Mutation, new chromosome{i}: {chromosome_final}') 
    
    print('\n\n\n')
    
```
